# 1 Membuat kolom kondisional dari 2 atau lebih pilihan np.select()

```python
import numpy as np
```

```python
conditions = [
    dfl['Price_euros']>3000,
    (dfl['Price_euros']>2000) & (dfl['Price_euros']<=3000),
    (dfl['Price_euros']>800) & (dfl['Price_euros']<=2000),
    dfl['Price_euros']<=800,
]

values = ['Too Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
# set ke kolom baru
dfl['Price_tiers']=np.select(conditions, values)
dfl
```

```python
dfl['Price_tiers'].value_counts()
```

```python
conditions_screen = [
    dfl['Inches']>16,
    (dfl['Inches']>14) & (dfl['Inches']<=16),
    (dfl['Inches']>12) & (dfl['Inches']<=14),
    dfl['Inches']<=12,
]

values_screen = ['Too Big', 'Big', 'Small', 'Too Small']
```

```python
dfl['Screen_size']=np.select(conditions_screen, values_screen)
dfl
```

```python
dfl['Screen_size'].value_counts()
```
