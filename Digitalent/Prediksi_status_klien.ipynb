```python
import pandas as pd 
df = pd.read_csv('https://raw.githubusercontent.com/smirzap/AI-dengan-DL/refs/heads/main/data/BankChurners.csv')
df
```

```python
df.info()
```

```python
df.rename(columns={'Attrition_Flag':'status'}, inplace=True)
df['status'].value_counts()
```

```python
#ambil pelanggan-pelanggan dengan status keluar
df_keluar = df[df['status']=='Attrited Customer']
dfk = df['status']=='Attrited Customer'
#ambil pelanggan-pelanggan dengan status setia
df_setia = df[df['status']=='Existing Customer']
```

```python
df_keluar_sampled = df_keluar.sample(400, random_state=2)
```

```python
dataset = pd.concat([df_setia, df_keluar_sampled])
```

```python
dataset['status'].value_counts()
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import recall_score
```

```python
# Hapus kolom status sehingga yang tersisa hanya data pokok
data_pokok = dataset.drop('status', axis=1)

# Ambil kolom status dari dataset
data_label = dataset['status']
```

```python
data_pokok.info()
```

```python
df_keluar
```

```python
dfk
```

```python
data_pokok = pd.get_dummies(data_pokok, drop_first=True)
```

```python
data_pokok.info()
```

```python
data_pokok_train, data_pokok_test, data_label_train, data_label_test = train_test_split(
data_pokok, data_label,
test_size=0.3, random_state=2)
```

```python
data_pokok_train, data_pokok_test, data_label_train, data_label_test
```

```python
model = DecisionTreeClassifier(max_depth=2,random_state=2)
model.fit(data_pokok_train, data_label_train)
```

```python
from sklearn import tree
rules = tree.export_text(model)
print(rules)
```

```python
#prediksi status masing-masing pelanggan pada test set
prediksi = model.predict(data_pokok_test)
#hitung nilai Recal dari hasil prediksi
recall_score(data_label_test, prediksi, average=None)

print(prediksi)
print(recall_score)
```
