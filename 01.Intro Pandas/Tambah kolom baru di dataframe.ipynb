# 2 Kolom baru dengan cara assign() dan insert()

Assign digunakan saat:
    - ingin menambahkan beberapa kolom di satu line kode
    - ketika ingin mengubah value dari kolom yang sudah ada
nanti akan mengembalikan objek baru dengan semua kolom original ke yang baru

```python
df1['score1'] = 70
```

```python
df1
```

```python
score1 = np.random.randint(1,100, size=2194488)
```

```python
series1 = pd.Series(score1, index=np.arange(0, 2194488))
```

```python
df1 = df1.assign(score1 = series1)
```

```python
df1
```

```python
df1.insert(1, 'score2', series1)
```
