```python
import pandas as pd
```

```python
dfl = pd.read_csv('Laptop_price1.csv')
dfl
```

# 1 drop_duplicates()

```python
# drop duplicates in 2 or more columns
dfl.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
# sort dataframe asending by company and price
# termurah dlu baru termahal
dfl = dfl.sort_values(['Company', 'Price_euros'])
```

```python
dfl.drop_duplicates(['Company'], keep='first') [['Company', 'Price_euros']]
```

```python
dfl.drop_duplicates(['Company'], keep='last', inplace=True, ignore_index=True)
dfl
```

```python
dfl[['Company', 'Price_euros']]
```
