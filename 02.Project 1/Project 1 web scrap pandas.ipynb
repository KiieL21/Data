```python
for season in range(20, 25):
    print(str(season) + str(season+1))
```

```python
leagues = ['E2', 'E3']
frames = []
for league in leagues:
    for season in range(20, 25):
        df = pd.read_csv( root + str(season) + str(season+1) + '/'+ league)
        df.insert(1, 'season', season)
        frames.append(df)
```

# 3 Organize all the data in a dict

```python
dict_countries = {
    'Spanish La Liga': 'SP1', 'Spanish Segunda Division':'SP2', 'German Bundesliga': 'D1', 'English Premier League':'E0', 'English League 1':'E2'
}
```

```python
dict_countries['English Premier League']
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {}

for league in dict_countries:
    frames = []
    for season in range(20, 25):
        df = pd.read_csv( root + str(season) + str(season+1) + '/'+ dict_countries[league] )
        df.insert(1, 'season', season)
        frames.append(df)
    df_con = pd.concat(frames)
    dict_historical_data[league] = df_con
```
