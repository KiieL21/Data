```python
# Example 2
# sort ascending berdasarkan Perusahaan dan Harga
dfl = dfl.sort_values(['Company','Price_euros'])
dfl
```

```python
dfl.value_counts('Company')
```

```python
dupe_fst = dfl.duplicated('Company', keep='first')
```

```python
dfl[dupe_fst]

dfl[~dupe_fst] [['Company', 'Price_euros']]
```

```python
dfl[~dupe_fst].value_counts('Company')
```

```python
dupe_last = dfl.duplicated('Company', keep='last')
dupe_false = dfl.duplicated('Company', keep=False)
dfl[~dupe_last][['Company', 'Price_euros']]
```

```python
dfl[~dupe_false]
```

```python
dfl.value_counts('Company')
```
