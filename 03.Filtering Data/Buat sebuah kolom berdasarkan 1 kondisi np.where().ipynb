```python
dfl = pd.read_csv('laptop_price1.csv')
dfl.head(3)
```

# 1.Buat kolom 1 kondisiLnp.where()

```python
import numpy as np
```

```python
# Buat array berdasarkan tingkatan harga
np.where(dfl['Price_euros']>2000, 'Expensive', 'Cheap')
```

```python
# set ke kolom baru
dfl['Price_tier'] = np.where(dfl['Price_euros']>2000, 'Expensive', 'Cheap')
```

```python
dfl.head()
```

```python
dfl.value_counts('Price_tier')
```
